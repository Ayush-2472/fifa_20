```python
df = df[['age','overall']].head()
df.head()
```

```python
df.dropna(subset=['age', 'overall'])
```

```python
# Group by age and calculate average overall rating
age_rating = df.groupby('age')['overall'].mean().reset_index()
```

```python
# Plot the distribution
plt.figure(figsize=(12, 6))
sns.lineplot(data=age_rating, x='age', y='overall', marker='o')
plt.title('Average Overall Rating vs Age of FIFA 20 Players')
plt.xlabel('Age')
plt.ylabel('Average Overall Rating')
plt.grid(True)
plt.show()
```

# From the above graph we can see that the players improvment slows down after the age of 32

```python
df = pd.read_csv('players_20.csv')
df.head()
```

```python
df.drop(columns=['sofifa_id','player_url'])
```

```python
df.describe().columns.T
```

```python
df = df[['player_positions','wage_eur']].head()
df.head()
```

```python
#Filter for offensive position
offensive_positions = ['ST','LW','RW']
offensive_df = df[df['player_positions'].isin(offensive_positions)]
```

```python
#Groupby position and calculate average wage
average_position = offensive_df.groupby('player_positions')['wage_eur'].mean().sort_values(ascending = False)
```

```python
#Plotting
plt.figure(figsize=(8, 6))
sns.barplot(x=average_position.index, y = average_position.values, palette ='coolwarm')
plt.title('Average wage (EUR) by players postion')
plt.xlabel('player_positions')
plt.ylabel('wage_eur')
plt.show()
```

# From above graph we can see that the Striker tends to get paid more
